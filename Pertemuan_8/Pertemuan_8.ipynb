Graph

Representasi Graph dengan Adjacency List

```python
# membuat graf tak berarah menggunakan adjacency list

graph = {
    'A' : ['B', 'C'],
    'B' : ['A', 'D', 'E'],
    'C' : ['A', 'F'],
    'D' : ['B'],
    'E' : ['B', 'F'],
    'F' : ['C', 'E']
}

# menampilkan graf
def print_graph(graf):
  for node in graf:
    print(f"{node} -> {graf[node]}")

print_graph(graph)
```

Traversal Graf menggunakan BFS

```python
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])
# contoh pemanggilan
bfs(graph, 'A')
```

Traversal Graf menggunakan DFS

```python
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)
# contoh pemanggilan
dfs(graph, 'A')
```

```python

graph = {
    'A' : ['B', 'C'],
    'B' : ['A', 'D', 'E'],
    'C' : ['A', 'F'],
    'D' : ['B'],
    'E' : ['B', 'F'],
    'F' : ['C', 'E']
}

def count_neighbors(graph):
  result = {}

  for node in graph:
    result[node] = len(graph[node])

  return result
# contoh pemanggilan
hasil = count_neighbors(graph)

for node in hasil:
  print(node, ":", hasil[node])
```

```python
# Implementasikan BFS dan DFS dari simpul awal pilihan anda

from collections import deque

# BFS
def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()

    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend(graph[vertex])

# DFS
def dfs(graph, start, visited=None):
  if visited is None:
    visited = set()

  if start not in visited:
    print(start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs(graph, neighbor, visited)

print("BFS: ")
bfs(graph, 'A')

print("\n\nDFS: ")
dfs(graph, 'A')
```

```python
# Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan, bukan hanya mencetak

from collections import deque


graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['C', 'D']
}

# BFS mengembalikan list
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    hasil = []

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            visited.add(vertex)
            hasil.append(vertex)

            queue.extend(graph[vertex])

    return hasil

print(bfs(graph, 'A'))
```

```python
# Buat fungsifind_path(graph, start, end) yang mengembalikan satu jalur dari start ke end jika ada, menggunakan DFS

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['C', 'D']
}

# Fungsi mencari jalur menggunakan DFS
def find_path(graph, start, end, visited=None):
    if visited is None:
        visited = []

    visited.append(start)

    if start == end:
        return visited

    for neighbor in graph[start]:
        if neighbor not in visited:
            hasil = find_path(graph, neighbor, end, visited)

            if hasil:
                return hasil

    return None

print(find_path(graph, 'A', 'E'))
```

```python
# Tambahkan fungsi is_connected(graph) untuk mencetak apakah semua simpul dalam graf saling terhubung(gunakan kombinasi DFS = logika)

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['C', 'D']
}

# DFS
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

    return visited

# Mengecek graph terhubung
def is_connected(graph):
    visited = dfs(graph, 'A')

    if len(visited) == len(graph):
        return True
    else:
        return False

print(is_connected(graph))
```
