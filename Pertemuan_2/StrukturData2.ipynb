Hapus Node dengan Kata Kunci

```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

# print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

# Function to add node in the begining
  def AddBegining(self, newdata):
    NewNode = Node(newdata)
#update the new nods nextval to existing node
    NewNode.nextval = self.headval
    self.headval = NewNode

# Function to remove node by key
  def RemoveNode(self, Removekey):
    HeadVal = self.headval
    if (HeadVal is not None):
      if (HeadVal.dataval == Removekey):
        self.headval = HeadVal.nextval
        HeadVal = None
        return
    while (HeadVal is not None):
      if HeadVal.dataval == Removekey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval
    if (HeadVal == None):
      return
    prev.nextval = HeadVal.nextval
    HeadVal = None

# create linked list named Li
Li = LinkedList()
# add node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")
print("Sebelum dilakukan remove")
Li.listprint()
print("========================================")

# remove node by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

Double linked list = 3 node/ penyisipan tengah dan hapus dengan kata kunci

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" --> ")
            temp = temp.next
        print("None")

    def add_end(self, data):
        new = Node(data)
        if self.head is None:
            self.head = new
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new
        new.prev = temp

    # Sisip di tengah (setelah node kedua)
    def insert_middle(self, data):
        temp = self.head.next  # langsung ke node kedua
        new = Node(data)

        new.next = temp.next
        new.prev = temp

        if temp.next:
            temp.next.prev = new

        temp.next = new

    # Hapus berdasarkan kata kunci
    def delete(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                if temp.prev:
                    temp.prev.next = temp.next
                else:
                    self.head = temp.next

                if temp.next:
                    temp.next.prev = temp.prev
                return
            temp = temp.next


# ===== Program Utama =====
dll = DoubleLinkedList()

# 3 node awal
dll.add_end("Januari")
dll.add_end("Februari")
dll.add_end("April")

print("Sebelum penyisipan :")
dll.display()

dll.insert_middle("Maret")

print("Setelah sisip tengah :")
dll.display()

dll.delete("Maret")

print("Setelah hapus 'Maret':")
dll.display()
```

Circular linked list dengan 3 node = penyisipan akhir dan hapus akhir

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        if self.head is None:
            return

        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(kembali ke head)")

    def add_end(self, data):
        new = Node(data)

        if self.head is None:
            self.head = new
            new.next = self.head
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new
        new.next = self.head

    def remove_end(self):
        if self.head is None:
            return

        if self.head.next == self.head:
            self.head = None
            return

        temp = self.head
        while temp.next.next != self.head:
            temp = temp.next

        temp.next = self.head


# ===== Program Utama =====
cll = CircularLinkedList()

# 3 node awal
cll.add_end("Apel")
cll.add_end("Mangga")
cll.add_end("Jeruk")

print("Sebelum tambah akhir:")
cll.display()

cll.add_end("Alpukat")

print("Setelah tambah akhir:")
cll.display()

cll.remove_end()

print("Setelah hapus akhir:")
cll.display()
```
